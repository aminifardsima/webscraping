```python
from itertools import groupby

def split_list_by_empty_string(input_list):
    # Use groupby to group items by empty strings
    grouped = groupby(input_list, lambda x: x == '')
    
    # Use a list comprehension to filter out empty groups and create sublists
    result = [list(group) for is_empty, group in grouped if not is_empty]
    
    return result
```

```python
result=split_list_by_empty_string(input_list)[1:7]
```

```python
list_of_columns=list(filter(None, result))
```
